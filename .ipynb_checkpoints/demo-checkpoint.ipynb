## 2. Imports

```python
from datasets import load_dataset
from transformers import AutoTokenizer, AutoModelForCausalLM
from peft import LoraConfig, get_peft_model
import torch
```

## 3. Peek at the dataset
We use TinyStories (default) or WikiText-2. Load TinyStories and inspect a couple rows to understand the style.

```python
dataset = load_dataset("roneneldan/TinyStories")
print(dataset)
samples = dataset['train'][:3]['text']
for i, text in enumerate(samples, 1):
    print(f'--- Sample {i} ---')
    print(text[:400])
    print()
```

## 4. Tokenizer & tokenized batches
Tokenize text with `distilgpt2`'s tokenizer and chunk into 512-token blocks (matching `train.py`).

```python
tokenizer = AutoTokenizer.from_pretrained("distilgpt2")
tokenizer.pad_token = tokenizer.eos_token
encoded = tokenizer(samples, padding=True, truncation=True, max_length=64, return_tensors='pt')
print('Input IDs shape:', encoded['input_ids'].shape)
print('First row token ids:', encoded['input_ids'][0][:20])
print('Decoded back:', tokenizer.decode(encoded['input_ids'][0]))
```

## 5. Load the base model
`distilgpt2` is a 6-layer GPT-2 distilled model. Let's inspect parameter counts and a sample generation.

```python
model = AutoModelForCausalLM.from_pretrained("distilgpt2")
total_params = sum(p.numel() for p in model.parameters())
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f'Total parameters: {total_params/1e6:.2f}M (trainable: {trainable_params/1e6:.2f}M)')
inputs = tokenizer("Tell me a story about robots in space.", return_tensors='pt')
output = model.generate(**inputs, max_new_tokens=50)
print(tokenizer.decode(output[0], skip_special_tokens=True))
```

## 6. Compare LoRA adaptation
LoRA only trains a small set of rank-decomposed adapters. This cell mirrors the setup in `src/train.py` and reports how many parameters remain trainable after wrapping the base model.

```python
lora_config = LoraConfig(
    r=16,
    lora_alpha=32,
    target_modules=["c_attn", "c_proj"],
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM",
)
lora_model = get_peft_model(model, lora_config)
lora_trainable = sum(p.numel() for p in lora_model.parameters() if p.requires_grad)
print(f'LoRA trainable params: {lora_trainable/1e6:.2f}M ({lora_trainable/total_params*100:.2f}% of full model)')
```

## Butterfly Adapters

- `--mode butterfly` inserts a full-width butterfly rotation in front of every projection listed in `--target_modules` (default `c_attn`, `c_proj`). The rotations are factored into staged 2×2 mixes so you train only ~O(d log d) angles while the original weights stay frozen.
- `--mode butterfly_block` (new) uses three independent 256-dim rotations for DistilGPT-2’s 768 channels. Pass `--butterfly_block_size 256 --butterfly_stages 8` for the clean log-depth configuration discussed in the notebook.
- Both modes work just like LoRA in the CLI: point them to your dataset, block size, etc., and they’ll report trainable parameter counts (~80k for full-width, ~60k for blockwise by default).
- To inspect what’s happening, run `python -m src.train --mode butterfly_block ... --logging_steps 10` and check `outputs/.../log_history_full.json` for eval loss / perplexity.


### Blockwise Butterfly Math

For `--mode butterfly_block` we treat the hidden vector $x \in \mathbb{R}^d$ as a stack of contiguous blocks of size $b$ (default 256):
\[
$ x = \begin{bmatrix} x^{(0)} \\ x^{(1)} \\ \vdots \\ x^{(m-1)} \end{bmatrix}, \qquad x^{(k)} \in \mathbb{R}^{b_k},\; b_k = \min(b, d - k b$).
\]
Each block gets its own butterfly rotation $R^{(k)}$, so the overall transform is block diagonal
\[
$R = \mathrm{diag}\big( R^{(0)}, R^{(1)}, \dots, R^{(m-1)} \big)$.
\]

The stage structure inside a block of length $L$ pairs indices according to the stride $2^s$:
\[
$\mathcal{P}_s = \left\{ (i, j) \mid i = \ell + r,\ j = i + 2^s,\ \ell \in \{0, 2^{s+1}, 2 \cdot 2^{s+1}, \dots\},\ 0 \le r < 2^s,\ j < L \right\}$.
\]
Each pair stores a learnable angle $\theta_{s,(i,j)}$ and applies a $2\times2$ rotation
\[
Q(\theta) = \begin{bmatrix}
\cos \theta & -\sin \theta \\
\sin \theta & \cos \theta
\end{bmatrix}.
\]
Because every stage matrix is orthogonal ($Q^\top Q = I$), the product of stages is orthogonal as well, so norms are preserved. Even though blocks are independent inside $R$, the frozen projection $W$ that follows is dense, so the effective weight $W R$ still mixes channels across blocks while benefiting from the clean log-depth butterfly inside each block.
